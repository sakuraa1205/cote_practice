### Chapter 5 DFS / BFS
- 탐색 알고리즘
- 스택, 큐, 재귀 함수의 이해가 전제되어야 한다.  
    - 스택 : 박스 쌓기에 비유 / 입구와 출구가 한 곳 뿐이다. / 별도의 라이브러리가 필요없고, 기본 리스트에서 append()와 pop() 사용  
    - 큐 : 놀이공원 대기줄에 비유 / 입구와 출구가 따로 있다. / collections 모듈에서 제공하는 deque 라이브러리 사용  
    - 디큐(deque) : 스택과 큐의 장점을 모두 채택 / 리스트보다 빠르고 효율적  
    - 재귀 함수 : 자기 자신을 다시 호출하는 함수 / 반복문보다 더 간결하게 작성가능
  
  
- 스택과 큐의 핵심
    - 삽입(Push) : 데이터를 삽입한다.  
    - 삭제(Pop) : 데이터를 삭제한다.  
    - 오버플로(Overflow) : 데이터의 크기가 가득 찬 상태에서 삽입연산을 수행할 때 발생한다.  
    - 언더플로(Underflow) : 데이터가 전혀 들어있지 않은 상태에서 삭제연산을 수행할 때 발생한다.  

```python
# 스택 예제
stack = []

stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
stack.pop()
stack.append(1)
stack.append(4)
stack.pop()

print(stack) # 최하단 원소부터 출력 (먼저 들어간 원소부터 출력)
print(stack[: : -1]) # 최상단 원소부터 출력 (늦게 들어간 원소부터 출력)
```

```python
# 큐 에제
from collections import deque

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue)
queue.reverse()
print(queue)
print(list(queue))
```

```python
# 재귀 함수 예제 2
def recursive_function(i):
    if i == 100:
        return
    print(i, '번째 재귀 함수에서', i + 1, '번째 재귀 함수를 호출합니다.')
    recursive_function(i + 1)
    print(i, '번째 재귀 함수를 종료합니다.')

recursive_function(1)
```

```python
# 팩토리얼 예제
# 반복적으로 구현한 n!
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)

print('반복적으로 구현:', factorial_iterative(5))
print('재귀적으로 구현:', factorial_recursive(5))
```

#### DFS (깊이 우선 탐색)
- 인접 행렬 : 2차원 배열로 그래프의 연결 관계를 표현하는 방식
- 인접 리스트 : 리스트로 그래프의 연결 관계를 표현하는 방식
- 스택 / 재귀 함수 이용

```python
# 인접 행렬 방식 예제
INF = 999999999 # 무한의 비용 선언

graph = [
    [0, 7, 5],
    [7, 0, INF],
    [5, INF, 0]
]

print(graph)
```

```python
# 인접 리스트 방식 예제
graph = [[] for _ in range(3)]

graph[0].append((1, 7))
graph[0].append((2, 5))

graph[1].append((0, 7))

graph[2].append((0, 5))

print(graph)
```

```python
# DFS 예제
def dfs(graph, v, visited):
    visited[v] = True
    print(v, end = ' ')
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
    
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * 9

dfs(graph, 1 , visited)
```

#### BFS (너비 우선 탐색)
- 가까운 노드부터 탐색
- 코딩 테스트 단계에서는 보통 BFS가 DFS보다 조금 더 빠르게 동작한다.
- 큐 / 큐 자료 이용

```python
# BFS 예제
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True
    
    while queue:
        v = queue.popleft()
        print(v, end = ' ')
        
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * 9

bfs(graph, 1 , visited)
```

#### 음료수 얼려 먹기

```python
# 교재 답안
n, m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int,input())))

def dfs(x, y):
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False
    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs(x - 1, y)
        dfs(x, y - 1)
        dfs(x + 1, y)
        dfs(x, y + 1)
        return True
    return False

result = 0
for i in range(n):
    for j in range(m):
        if dfs(i, j) == True:
            result += 1
    
print(result)
```

#### 미로 탈출

```python
# 교재 답안
from collections import deque

n, m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int, input())))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
    
def bfs(x,y):

    queue = deque()
    queue.append((x,y))
    
    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if nx >= n or ny >= m or nx < 0 or ny < 0 :
                continue
            
            if graph[nx][ny] == 0:
                continue
            
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1
                queue.append((nx,ny))
                
    return graph[n-1][m-1]

print(bfs(0,0))
```
